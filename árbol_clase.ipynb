## **ACTIVIDAD**

## 1.Tomar el dataset de advertising y hacer una regresion lineal simple

```python
dataset = pd.read_csv('Advertising.csv')
dataset.head()
```

```python
# definir Y y X
y = dataset['sales']
x = dataset[['TV','radio','newspaper']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

## 2. Calcular R2

```python
R2 = lr.score(x, y)
R2
```

## 3.Usar 1 árbol de decisión para regresión sobre el mismo dataset

```python
from sklearn.tree import DecisionTreeRegressor
DecisionTreeRegressor = DecisionTreeRegressor()
DecisionTreeRegressor.fit(x, y)
```

## 4. Calcular R2

```python
R2_tree = DecisionTreeRegressor.score(x, y)
R2_tree
```

## 5.Repetir ( 1- 4) con train-test al 50/50

```python
# Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)
```

### *Con árboles*

```python
# Árboles con train / test split
from sklearn.tree import DecisionTreeRegressor
DecisionTreeRegressor50 = DecisionTreeRegressor()
DecisionTreeRegressor50.fit(X_train, y_train)
```

```python
# R2 en Train con árbol
R2_tree_train = DecisionTreeRegressor50.score(X_train, y_train)
R2_tree_train
```

```python
# R2 en Test con árbol
R2_tree_test = DecisionTreeRegressor50.score(X_test, y_test)
R2_tree_test
```

### *Con regresión*

```python
# Hacemos regresión con Train/Test
lr2 = LinearRegression()
lr2.fit(X_train, y_train)
```

```python
# Calculamos el R2 de train
R2_train = lr2.score(X_train, y_train)
R2_train
```

```python
# Calculamos el R2 de test
R2_test = lr2.score(X_test, y_test)
R2_test
```
