```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

- Carga de datos

```python
data = pd.read_excel('Boston Housing Dataset 1978.xlsx')
```

```python
data.tail()
```

- EDA

```python
data.describe()
```

```python
# Es el precio promedio de la casa
y = data['MEDV']
```

- Escalamiento min max

```python
# todas las variables MENOS LA Y
x = data.drop('MEDV', axis=1)
x = (x - x.min()) / (x.max() - x.min())
x.describe()
```

```python
columnas = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

- Escalamiento estandar

```python
# le resto media y dividio entre desviacion
x = (x - x.mean()) / x.std()
x.describe()
```

```python
plt.hist(x['CRIM'],  bins=20, alpha=0.5)
plt.hist(x['ZN'],  bins=20, alpha=0.5)
plt.hist(x['INDUS'],  bins=20, alpha=0.5)
plt.hist(x['CHAS'],  bins=20, alpha=0.5)
plt.hist(x['NOX'],  bins=20, alpha=0.5)
plt.hist(x['RM'],  bins=20, alpha=0.5)
plt.hist(x['AGE'],  bins=20, alpha=0.5)
plt.hist(x['DIS'],  bins=20, alpha=0.5)
plt.hist(x['RAD'],  bins=20, alpha=0.5)
plt.hist(x['TAX'],  bins=20, alpha=0.5)
plt.hist(x['PTRATIO'],  bins=20, alpha=0.5)
plt.hist(x['B'],  bins=20, alpha=0.5)
plt.hist(x['LSTAT'],  bins=20, alpha=0.5)
plt.legend(loc="best")
plt.ylabel('MEDV')
plt.title('Histogram')
plt.show()
```
