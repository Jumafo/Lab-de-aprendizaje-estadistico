```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Salarios

```python
salarios = pd.read_csv('Salarios_minimos.csv')
salarios.head()
# x es la variable independiente = lo que puedo controlar
# x = periodo
# y es la variable dependiente = NO PUEDO CONTROLAR
# y = salario
```

```python
salarios.tail()
```

```python
# extraemos la columna que necesitamos
x = salarios['Periodo']
# reshape = Forma: (n, 1) - array 2D (matriz columna)
x = np.array(x).reshape(-1, 1)
```

```python
y = salarios['Salario']
y = np.array(y).reshape(-1, 1)
```

**- Realiza regresiones polinomiales de orden 1, 3, 5 y 7**

#####***ORDEN 1***

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Nombramos las variables para que sea más fácil graficarlas
x_b0 = lr.intercept_
x_b1 = lr.coef_
```

Para la regresion lineal de orden 1, $\beta_0$ = -5022.60 y $\beta_1$ = 2.5283

```python
# R2 es la varianza que SE EXPLICA
R2_orden1 = lr.score(x, y)
R2_orden1

# para obtener la varianza que NO SE EXPLICA en porcentaje
(1-R2_orden1)
```

#####***ORDEN 3***

```python
x2 = x*x
x2
```

```python
x3 = x2*x
x3
```

```python
# matriz
# hstack = que apile los vectores en horizontal
X =  np.hstack((x, x2, x3))
X
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# sirve para hacer predicciones usando el modelo que ya entrenaste
# va a servir para hacer la grafica
# es como si hicieramos esto y_orden_3 = (x3_b0 + x3_b1 * x + x3_b2 * x2 + x3_b3 * x3)
y_orden_3 = lr.predict(X)
y_orden_3
```

```python
x3_b0 = lr.intercept_

# RECUERDA que lr.coef es UNA fila con TRES columnas
x3_b1 = lr.coef_[0][0]

# por eso el primer es indice, das el num de la fila y el segundo el numero de columnas
x3_b2 = lr.coef_[0][1]
x3_b3 = lr.coef_[0][2]
```

Para la regresion de orden 3:

*   $\beta_0$ = -5217722.95
*   $\beta_1$ = 7.83146
*   $\beta_2$ = -3.9193
*   $\beta_3$ = 0.0006





```python
# R2 es la varianza que SE EXPLICA
R2_orden3 = lr.score(X, y)
R2_orden3

# para obtener la varianza que NO SE EXPLICA en porcentaje
(1-R2_orden3)
```

*   $\beta_5$ = 1.02560835e-10
*   $\beta_6$ = -8.54813519e-14
*   $\beta_7$ = 1.83257513e-17



```python
# R2 es la varianza que SE EXPLICA
R2_orden7 = lr.score(Xfinal, y)
R2_orden7

# para obtener la varianza que NO SE EXPLICA en porcentaje
(1-R2_orden7)
```

**- Compara el R2 que te da cada modelo.**

```python
# R2 es la varianza que SE EXPLICA
# index = tus columnas
# .T = transponer la tabla

R2_table = pd.DataFrame(index = ['Orden', 'R2'],
                        data = [['1', '3', '5', '7'], [R2_orden1 , R2_orden3, R2_orden5, R2_orden7]])
R2_table.T
```

**- Crea una gráfica con una nube de dispersión (datos originales) y tus modelos**

```python
# Crear figura
plt.figure(figsize=(14, 8))

# PREDICCIÓN ORDEN 1
y_orden_1 = x_b0 + x_b1 * x

# Graficar las regresiones
#  .4f = 4 decimales
plt.plot(x, y_orden_1, 'o-', label=f'Orden 1 (R²={R2_orden1:.4f})', linewidth=2.5)
plt.plot(x, y_orden_3, 's-', label=f'Orden 3 (R²={R2_orden3:.4f})', linewidth=2.5)
plt.plot(x, y_orden_5, '^-', label=f'Orden 5 (R²={R2_orden5:.4f})', linewidth=2.5)
plt.plot(x, y_orden_7, 'd-', label=f'Orden 7 (R²={R2_orden7:.4f})', linewidth=2.5)

# Graficar las observaciones
# zorder=10 es para que queden encima
plt.scatter(x, y, color='black', label='Observaciones', s=100, zorder=10)

# Configuración de la gráfica
plt.xlabel('Periodo en años')
plt.ylabel('Salario Mínimo')
plt.title('Regresión Polinomial - Salarios Mínimos (1988-2018)')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

A partir de esta gráfica se puede concluir que los salarios minimos han aumentado de manera lineal durante 30 años. Al comparar los modelos se puede observar que se sobreponen, no hay mucha diferencia entre ellos, por lo que se recomienda utilizar el menos complejo; es decir, como todos los modelos se sobreponen no hay necesidad de usar el modelo más complejo, puesto que el más sencillo obtendrá resultados muy similares.

**- Crea una gráfica con una nube de dispersión (datos originales) y tus modelos**

```python
# Tenemos que ordenar los datos antes de graficar
# en la de salarios, estaban ordenados por años y de manera consecutiva pero los de batería estan desordenados y si los dejamos así, los puntos no van a formar una línea

# xb.flatten es para aplanar o poner todo en una misma dimensión
# xb = [[1], [5], [2], [8]]  (matriz con forma (4,1))
# con xb.flatten() el resultado: [1, 5, 2, 8] (vector simple)
xb_aplanado = xb.flatten()

# Obtener índices de ordenamiento
# np.argsort me dice cómo ordenar los datos, devuelve las POSICIONES (índices) que ordenarían el array
# ambos te permiten ordenar tus datos manteniendo la relación x-y correcta para que la gráfica se vea bien
indices_ordenados = np.argsort(xb_aplanado)

# Ahora reorganizamos el array según los índices que vienen en el argsort, así que el resultado queda ordenado de menor a mayor
# porque es el orden más adecuado para mantener la relación x-y correcta para que la gráfica se vea bien ( como se menciona en las líneas de arriba)
xb_ordenado = xb_aplanado[indices_ordenados]

# Ordenar las predicciones
# yb_orden_1 = lr.predict(XB)  YA es 1D y tiene forma de un vector simple y ya está aplanado NO necesita .flatten()
# reorganiza las predicciones en el mismo orden que xb para que cada punto (x, y) siga siendo el par correcto
yb_orden_1_ordenado = yb_orden_1[indices_ordenados]
yb_orden_3_ordenado = yb_orden_3[indices_ordenados]
yb_orden_5_ordenado = yb_orden_5[indices_ordenados]
yb_orden_7_ordenado = yb_orden_7[indices_ordenados]

# Tamaño de la figura
plt.figure(figsize=(14, 8))

# graficar las regresiones - USAR LOS DATOS ORDENADOS
plt.plot(xb_ordenado, yb_orden_1_ordenado, '-', label=f'Orden 1 (R²={R2_orden1b:.4f})', linewidth=2.5)
plt.plot(xb_ordenado, yb_orden_3_ordenado, '-', label=f'Orden 3 (R²={R2_orden3b:.4f})', linewidth=2.5)
plt.plot(xb_ordenado, yb_orden_5_ordenado, '-', label=f'Orden 5 (R²={R2_orden5b:.4f})', linewidth=2.5)
plt.plot(xb_ordenado, yb_orden_7_ordenado, '-', label=f'Orden 7 (R²={R2_orden7b:.4f})', linewidth=2.5)

# graficar las observaciones
plt.scatter(xb, yb, color='black', label='Observaciones', s=100, zorder=10)

# Configuración de la gráfica
plt.xlabel('Tiempo de carga')
plt.ylabel('Duración de la batería')
plt.title('Regresión Polinomial - Duración de la batería')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

A partir de esta gráfica, se puede concluir que el tiempo de carga crece muy rápido y al llegar a las 8 horas, se estabiliza. Ahora bien, el modelo de orden 1 es el peor modelo; mientras que el 5 y 7 son los que mejor se ajustan a las observaciones, aunque el del orden 7 podría tender a sobreajustarse y es más complejo, por ende, el modelo de orden 5 es el más recomendable y un poco más sencillo que el de orden 7.
