```python
import pandas as pd
```

```python
data = pd.read_excel('Aguas del feo.xlsx')
data
```

```python
# Para saber como está tal cual y no tener errore en .map()
print(data['tamaño'].unique())
print(data['hielo'].unique())
```

```python
# Hacemos una copia del dataframe
data_new = data.copy()
```

```python
# .map() = reemplaza los valores de una columna usando un diccionario o una función
# osea cada que vea ESTO, lo cambie por ESTO
data_new['tamaño'] = data_new['tamaño'].map({'grande': 1, 'chica': 0})
data_new['hielo'] = data_new['hielo'].map({'con': 1, 'sin': 0})
data_new
```

```python
# Primero se crean las columnas
# Después si de la tabla de sabor, tienes kahlua te da TRUE y TRUE se interpreta como 1
# si NO es igual a kahlua, se interpreta como FALSE = 0
# astype cambia a boleanos

data_new['kahlua'] = (data_new['sabor'] == 'kahlua').astype(int)
data_new['horchata'] = (data_new['sabor'] == 'horchata').astype(int)
data_new
```

```python
# Eliminamos la columna de sabor
data_new = data_new.drop(columns=['sabor'])
data_new
```

```python
t = data_new['tamaño']
h = data_new['hielo']
k = data_new['kahlua']
ho = data_new['horchata']

X = data_new[["tamaño", "hielo", "kahlua", "horchata"]].values
y = data_new["score"]
```

```python
# Agregamos la columna de unos
import statsmodels.api as sm
X_unos = sm.add_constant(X)
X_unos
```

## Regresión - PARTE 1

```python
import statsmodels.api as sm
ols = sm.OLS(y, X_unos)
results = ols.fit()
# si dice uncentered = significa que no tiene la columna de unos
results.summary()
```

```python
beta_0 = results.params['const']
beta_0
```

## PARTE 3 - Comparar Beta 0

```python
table = pd.DataFrame(index = ['Parte', 'Beta 0'],
                        data = [['1','2'],
                         [beta_0, beta_02]])
table.T
```

A partir de los resultados obtenidos, se puede concluir que en la primera parte, cuando teníamos 1 y 0, el modelo tomaba 0 como un valor observado en la muestra que literalmente significa que NO HAY. Por otro lado, en la parte 2, 0 es un punto medio entre -1 y 1, por lo que ahora se toma como si fuera un promedio o algo teorico. Es decir, cuando tenemos 1 y 0, el punto de referencia se interpreta como HAY o NO HAY y cuando tenemos -1 y 1, el punto de referencia cambia y es más un promedio o tentativa. Además, es importante mencionar que la R2, predicciones y el modelo, se mantuvieron iguales en ambos casos.
