##**1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.**


```python
import pandas as pd
import numpy as np
```

#-EDA

```python
data = pd.read_excel('Motor Trend Car Road Tests.xlsx')
data.head()
```

```python
data.describe()
```

#- Definimos variables X y Y

```python
y = data['mpg']
y
```

```python
# hago mi matriz de X sin los boleanos para poder estandarizar
X = data.drop(['mpg', 'model', 'am', 'vs'], axis=1)
X
```

#-Estandarización

```python
# Estandarizamos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_estandarizar= scaler.transform(X)
```

```python
# Verficamos que Media = 0 y Desviacion estandar = 1
X_estandarizar = pd.DataFrame(X_estandarizar)
X_estandarizar.describe()
```

```python
# Lo hacemos en formato numpy para que se pueda concatenar con el data estadarizado
data[['vs', 'am']].values
```

```python
# Hacemos X matriz y le agregamos las columnas boleanas
X = np.hstack([X_estandarizar, data[['vs', 'am']].values])
X
```

#-Regresión sin penalización

```python
import statsmodels.api as sm
X_con_unos= sm.add_constant(X)
model = sm.OLS(y, X_con_unos)
results = model.fit()
results.summary()
```

#- Calcula el R2 e interpreta los signos de los betas

```python
r2 = results.rsquared
r2
```

A partir de los resultados obtenidos, las variables con efecto negativo son x1, x3, x5 y x8, esto significa que un incremento en estas varibles se vería reflejado en una disminución de millas por galón. El coeficiente que más llama la atención por ser el más negativo del modelo es el x5 = peso en toneladas con -3.5780, esto significa que entre más pesado sea el coche, el rendimiento de millas por galón disminuye; además, tiene un p-value de 0.063, muy cerca de ser considerado estadísticamente significativo. Por otro lado, las variables x2, x4, x6, x7, x9 y x10 presentan coeficientes positivos, destacando x10 = carb con el mayor coeficiente = 2.5202, sugieriendo que un incremento en esta variable, mejoraría el rendimiento del combustible.

Sin embargo, ninguno de los coeficientes muestran significancia estadística, ya que todos los p-values están por encima del 0.05; esto significa que a pesar de que el modelo tiene una R² alta (0.807), no hay suficiente evidencia estadística para afirmar que las variables individuales tengan un efecto real sobre la variable dependiente. Estos resultados sugieren que exite colinealidad entre las variables predictoras o que la habilidad predictiva del modelo surge a partir de la combinación de las varibles más que por el efecto individual de las variables.
